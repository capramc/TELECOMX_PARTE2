**Extra√ß√£o do Arquivo Tratado**

*Descri√ß√£o*

Carregue o arquivo CSV que cont√©m os dados tratados anteriormente.

üìÇ Aten√ß√£o: Utilize o mesmo arquivo que voc√™ limpou e organizou na parte 1 do desafio Telecom X. Ele deve conter somente as colunas relevantes, j√° com os dados corrigidos e padronizados.

```python
import pandas as pd
df = pd.read_csv('/content/TELECOM_.csv')
```

```python
# #Tratamento dos dados
# #--- MultipleLines
df["MultipleLines"] = df["MultipleLines"].replace("No phone service", "No_phone_service")

# #--- InternetService
df["InternetService"] = df["InternetService"].replace("Fiber optic", "Fiber_optic")

#--- OnlineSecurity
df["OnlineSecurity"] = df["OnlineSecurity"].replace("No internet service", "No_internet_service")

#--- OnlineBackup
df["OnlineBackup"] = df["OnlineBackup"].replace("No internet service", "No_internet_service")

#--- DeviceProtection
df["DeviceProtection"] = df["DeviceProtection"].replace("No internet service", "No_internet_service")

#--- TechSupport
df["TechSupport"] = df["TechSupport"].replace("No internet service", "No_internet_service")

#--- StreamingTV
df["StreamingTV"] = df["StreamingTV"].replace("No internet service", "No_internet_service")

#--- StreamingMovies
df["StreamingMovies"] = df["StreamingMovies"].replace("No internet service", "No_internet_service")

#--- Contract
df["Contract"] = df["Contract"].replace({"One year": "One_year","Two year": "Two_year", "Month-to-month": "Month_to_month"})

#--- PaymentMethod
df["PaymentMethod"] = df["PaymentMethod"].replace({
    "Mailed check": "Mailed_check",
    "Electronic check": "Electronic_check",
    "Credit card (automatic)": "Credit_card_(automatic)",
    "Bank transfer (automatic)": "Bank_transfer_(automatic)"
  })

#--- Charges.Monthly
df = df.rename(columns={"Charges.Monthly": "ChargesMonthly"})

#--- Charges.Total
df = df.rename(columns={"Charges.Total": "ChargesTotal"})

#--- Daily.Charge
df = df.rename(columns={"Daily.Charge": "DailyCharge"})

#--- tenure
df = df.rename(columns={"tenure": "Tenure"})

#--- gender
df = df.rename(columns={"gender": "Gender"})

#--- customerID
df = df.rename(columns={"customerID": "CustomerID"})
```

```python
import numpy as np

def calcular_largura_bin(df, coluna):
    Q75, Q25 = np.percentile(df[coluna], [75 ,25])
    IQR = Q75 - Q25
    largura_bin = 2 * IQR * np.power(len(df[coluna]), -1/3)
    return largura_bin
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def calcular_metricas_regressao(y_test, y_pred):
    # Calcula o RMSE manualmente tirando a raiz da MSE
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    metricas = {
        'Raiz do Erro Quadr√°tico M√©dio (RMSE)': round(rmse, 4),
        'Erro Absoluto M√©dio (MAE)': round(mae, 4),
        'R¬≤ (Coeficiente de Determina√ß√£o)': round(r2, 4)
    }

    return metricas
```

*Description*

```python
from imblearn.over_sampling import RandomOverSampler

# Oversampling (aumentar a classe minorit√°ria)
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y["Churn"])
```

```python
# Undersampling
arvore.score(X_resampled, y_resampled)
```

```python
from imblearn.over_sampling import SMOTE

# SMOTE (oversampling com dados sint√©ticos)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y["Churn"])
```

**Normalizando os dados**

Por exemplo, em uma base de dados com a colunas idade e sal√°rio, o algoritmo pode dar um peso de decis√£o maior para os valores do sal√°rio simplesmente por estar em uma escala maior do que os valores de idade, e n√£o porque a vari√°vel sal√°rio √© mais importante do que a vari√°vel idade.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def gerar_insights_correlacao(X, y, nome_alvo="Churn", mostrar_grafico=True):
    """
    Gera insights de correla√ß√£o positiva com a vari√°vel alvo.

    Par√¢metros:
    - X: DataFrame com vari√°veis explicativas (preferencialmente normalizadas)
    - y: Series ou DataFrame com a vari√°vel alvo (ex: 'Churn')
    - nome_alvo: nome da coluna da vari√°vel alvo
    - mostrar_grafico: se True, exibe gr√°fico de correla√ß√£o

    Retorno:
    - DataFrame com vari√°veis, correla√ß√£o e interpreta√ß√£o
    """

    # Garante que y seja Series com nome
    if isinstance(y, pd.DataFrame):
        y = y.iloc[:, 0]
    y.name = nome_alvo

    # Junta X e y para correla√ß√£o
    df_corr = pd.concat([X, y], axis=1)
    correlacao = df_corr.corr(numeric_only=True)[nome_alvo]
    cor_positiva = correlacao[correlacao > 0].sort_values(ascending=False).drop(nome_alvo)

    # Criar DataFrame de insights
    insights = pd.DataFrame({
        "Vari√°vel": cor_positiva.index,
        "Correla√ß√£o com " + nome_alvo: cor_positiva.values,
        "Interpreta√ß√£o (sugestiva)": ["A ser interpretada manualmente"] * len(cor_positiva)
    })

    # Mostrar gr√°fico
    if mostrar_grafico:
        plt.figure(figsize=(10, 6))
        sns.barplot(data=insights, y="Vari√°vel", x="Correla√ß√£o com " + nome_alvo, palette="Reds_r")
        plt.title(f"Vari√°veis com Maior Correla√ß√£o Positiva com {nome_alvo}")
        plt.xlabel("Correla√ß√£o")
        plt.ylabel("Vari√°vel")
        plt.tight_layout()
        plt.show()

    return insights
```

**O que cada m√©trica significa?**

    O quanto o modelo acerta no geral.

    F√≥rmula: (Verdadeiros Positivos + Verdadeiros Negativos) / Total

    Valor entre 0 e 1 (ou 0% e 100%)

    ‚ö†Ô∏è Pode ser enganosa se os dados estiverem desbalanceados (ex: muitos "n√£o churn").

üü¢ Exemplo Interpreta√ß√£o:
Uma acur√°cia de 0.85 (85%) significa que 85% das previs√µes foram corretas ‚Äî mas n√£o diz se ele est√° acertando os casos de churn ou s√≥ os de n√£o churn.


**Precis√£o (Precision)**

    Das vezes que o modelo previu "Churn", quantas estavam corretas?

    F√≥rmula: VP / (VP + FP)

    Foco em evitar falsos positivos.

üü¢ Exemplo Interpreta√ß√£o:
Uma precis√£o de 0.76 significa que 76% das vezes que o modelo disse que haveria churn, ele estava certo.

**Recall (Sensibilidade ou Revoca√ß√£o)**

    Das pessoas que realmente fizeram churn, quantas o modelo conseguiu identificar?

    F√≥rmula: VP / (VP + FN)

    Foco em n√£o deixar escapar positivos reais (churn reais).

üü† Exemplo Interpreta√ß√£o:
Um recall de 0.66 indica que o modelo detectou 66% de todos os churns reais. Os outros 34% ele perdeu.

**F1-score**

    Equil√≠brio entre Precis√£o e Recall.

    F√≥rmula: 2 * (Precis√£o * Recall) / (Precis√£o + Recall)

    Valor entre 0 e 1 ‚Äî quanto maior, melhor o equil√≠brio.

üü¢ Exemplo Interpreta√ß√£o:
Um F1-score de 0.70 indica um bom equil√≠brio entre acertar os churns sem acusar falsamente quem n√£o √©.


**Quando cada m√©trica √© mais importante?**

| Situa√ß√£o                       | M√©trica mais importante   | Por qu√™?                       |
| ------------------------------ | ------------------------- | ------------------------------ |
| Dados desbalanceados           | **F1-score** e **Recall** | Acur√°cia pode enganar.         |
| Custo de falso positivo √© alto | **Precis√£o**              | Ex: n√£o acusar cliente errado. |
| Custo de falso negativo √© alto | **Recall**                | Ex: n√£o deixar churn passar.   |
| Modelo equilibrado geral       | **F1-score**              | Resume bem os dois lados.      |


**Exemplo pr√°tico:**

Se voc√™ quer reduzir evas√£o (churn), o recall pode ser mais importante ‚Äî pois voc√™ n√£o quer deixar passar nenhum cliente que vai sair. Mas se voc√™ tamb√©m n√£o quer incomodar clientes fi√©is com alertas, precisa equilibrar com a precis√£o ‚Üí a√≠ o F1-score √© ideal para comparar modelos.
